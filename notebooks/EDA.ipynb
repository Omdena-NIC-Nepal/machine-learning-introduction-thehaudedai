```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline
```

```python
# Load the dataset
file_path = '../data/BostonHousing.csv'  # adjust if needed
df = pd.read_csv(file_path)
df
```

```python
# Dataset dimensions
print("Shape of dataset:", df.shape)

# Data types and non-null values
df.info()
```

```python
# Descriptive statistics
df.describe()
```

```python
# Count missing values per column
df.isnull().sum()
```

```python
# Count duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu', fmt='.2f', mask=np.triu(corr_matrix))
plt.title("Correlation Matrix")
plt.show()
```

```python
# Plot distribution of target variable
plt.figure(figsize=(8, 5))
sns.histplot(df['medv'], bins=30, kde=True)
plt.title("Distribution of Median House Prices (MEDV)")
plt.xlabel("MEDV ($1000s)")
plt.show()
```

```python
# Let's look at the top 3 correlated features with MEDV
top_corr = corr_matrix['medv'].abs().sort_values(ascending=False)[1:4].index.tolist()

# Plot scatterplots
for feature in top_corr:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=feature, y='medv')
    plt.title(f"{feature} vs MEDV")
    plt.show()
```

```python
# Boxplots to spot outliers in numeric features
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numeric_cols:
    plt.figure(figsize=(14, 2))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```
